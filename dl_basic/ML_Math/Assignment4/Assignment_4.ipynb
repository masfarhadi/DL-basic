<h2 style="color:rgb(0,120,170)">Task 1: Proof details for logistic regression</h2>


The coefficients $\mathbf{w}$ of the Logistic Regression problem are usually estimated using the maximum likelihood method.
Assuming that all the observations in the sample are independently Bernoulli distributed, this leads to the optimization problem

\begin{align*}
\min_{\mathbf{w}} L &= \min_{\mathbf{w}} \bigl( -\sum_i [ y_i \log \sigma(\mathbf{w}^T\mathbf{x}_i) + (1-y_i) \log (1-\sigma(\mathbf{w}^T \mathbf{x}_i))] \bigr), \\ 
\text{with} \quad \sigma(x) &= \frac{1}{1+\mathrm{e}^{-x}}, 
\end{align*}

where $L = -\ln \mathcal{L}$ is the negative log-likelihood. Since there is no closed-form solution for $\mathbf{w}$, an iterative method $-$ such as gradient descent $-$ must be applied in order to find $\mathbf{w}$.

<h3 style="color:rgb(0,120,170)">Task:</h3>

Calculate the gradient $\frac{\partial L}{\partial \mathbf{w}}$, which is needed to apply gradient descent.
Hint: $\frac{\partial \sigma(x)}{\partial x} = \sigma(x)\ (1 - \sigma(x))$.

<h3 style="color:rgb(210,90,80)">Calculation (25 points):</h3>

To calculate gradient of the negative log-liklihood 
\begin{align}
\frac{\partial L}{\partial\mathbf{w}} & =\frac{\partial L}{\partial\sigma}\frac{\partial\sigma}{\partial\mathbf{w}}=-\sum_{i}[\frac{\partial}{\partial\sigma(\mathbf{w}^{T}\mathbf{x}_{i})}[y_{i}\log\sigma(\mathbf{w}^{T}\mathbf{x}_{i})+(1-y_{i})\log(1-\sigma(\mathbf{w}^{T}\mathbf{x}_{i}))]\frac{\partial\sigma(\mathbf{w}^{T}\mathbf{x}_{i})}{\partial\mathbf{w}}]=\\
 & =-\sum_{i}[(y_{i}\frac{1}{\sigma(\mathbf{w}^{T}\mathbf{x}_{i})}-(1-y_{i})\frac{1}{1-\sigma(\mathbf{w}^{T}\mathbf{x}_{i})})\frac{\partial\sigma(\mathbf{w}^{T}\mathbf{x}_{i})}{\partial\mathbf{w}}]\\
 & =-\sum_{i}[\frac{(y_{i}-\sigma(\mathbf{w}^{T}\mathbf{x}_{i}))}{\sigma(\mathbf{w}^{T}\mathbf{x}_{i})(1-\sigma(\mathbf{w}^{T}\mathbf{x}_{i}))}\frac{\partial\sigma(\mathbf{w}^{T}\mathbf{x}_{i})}{\partial\mathbf{w}}]
\end{align}

by considering the hint,
\begin{align}
\frac{\partial\sigma(\mathbf{w}^{T}\mathbf{x}_{i})}{\partial\mathbf{w}}=\sigma(\mathbf{w}^{T}\mathbf{x}_{i})(1-\sigma(\mathbf{w}^{T}\mathbf{x}_{i}))\mathbf{x}_{i}
\end{align}

Therefore,
\begin{align}
\frac{\partial L}{\partial\mathbf{w}}=-\sum_{i}[\frac{(y_{i}-\sigma(\mathbf{w}^{T}\mathbf{x}_{i}))}{\sigma(\mathbf{w}^{T}\mathbf{x}_{i})(1-\sigma(\mathbf{w}^{T}\mathbf{x}_{i}))}]\sigma(\mathbf{w}^{T}\mathbf{x}_{i})(1-\sigma(\mathbf{w}^{T}\mathbf{x}_{i}))\mathbf{x}_{i}=-\sum_{i}[(y_{i}-\sigma(\mathbf{w}^{T}\mathbf{x}_{i}))\mathbf{x}_{i}]
\end{align}

<h3 style="color:rgb(210,90,80)">Code (20 points):</h3>

```python
import numpy as np
from sklearn.metrics import roc_curve, auc
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Complete the functions with your code


def logistic_gradient(w, x, y):
    ######################################
    # Begin your code
    sigmoid = lambda z:1/(1 + np.exp(-z))
    
    y = y.reshape(-1,1)
    z = np.dot(x, w)
    g = sigmoid(z)
    gradient = np.dot(x.T, (g- y)) 
    return gradient
    # End your code
    ######################################
    
def cost(w, x, y):
    ######################################
    # Begin your code
    sigmoid = lambda z:1/(1 + np.exp(-z))
    
    y = y.reshape(-1,1)
    z = np.dot(x, w)
    g = sigmoid(z)
    cross_ent_loss = - np.sum(y * np.log(g) + (1 - y) * np.log(1 - g))
    return cross_ent_loss
    # End your code
    ######################################
    
def numerical_gradient(w, x, y):
    ######################################
    # Begin your code
    eps = 1e-4
    num_gradient = np.zeros(w.shape)
    for ii in range(w.size):
        e = np.zeros(w.shape)
        e[ii]=eps
        num_gradient[ii] =(cost(w+e, x, y)-cost(w-e, x, y))/(2*eps)
    return num_gradient
    # End your code
    ######################################
```

```python
# Generate random data matrix and compare outputs

#################################################
# Begin your code

x = np.random.normal(size=(5, 10))
y = np.random.randint(2, size=5)
w = np.random.normal(size=(10, 1))
log_gradient = logistic_gradient(w, x, y)
print('logistic_gradient =',log_gradient.T)

num_gradient = numerical_gradient(w, x, y)
print('numerical_gradient =',num_gradient.T)

outputs_mse = np.mean((log_gradient-num_gradient)**2)
print('output MSE =', outputs_mse )


# End your code
#################################################
```

* Implement a function `fitLogRegModel(x_train, y_train, eta=1e-4, max_iter=100000)` that uses Logistic Regression with Gradient Descent to compute classifiers on the training set. Use randomly initialized weights, drawn from a uniform distribution between $-1$ and $1$, a learning rate $\eta$ of $10^{-4}$ and a maximum number of iterations of $100 000$. Furthermore the algorithm should stop if the difference between the original loss and the current loss is less than $\eta$. Also print the losses after $1000$ step intevals.
* Furthermore, implement a function `predictLogReg(w, x_pred)` that returns the prediction for the given parameter vector $\mathbf{w}$ and feature vector $\mathbf{x}_{\textrm{pred}}$.

Hint: for intialization use `np.random.uniform`.

<h3 style="color:rgb(208,90,80)">Code (20 points)</h3>

```python
# Complete the functions with your code

def fitLogRegModel(x_train, y_train, eta=1e-4, max_iter=100000):
    ######################################
    # Begin your code
    w = np.random.uniform(-1,1,size=(x_train.shape[1], 1))
    # Use of break statement inside the loop
    loss=[1]
    for cnt in range(max_iter):
        w = w - eta*logistic_gradient(w, x_train, y_train)
        loss.append(cost(w, x_train, y_train))
        if cnt % 1000 == 0:
            print( 'iter =', cnt, 'loss =', loss[-1])
        if np.abs(loss[-1]-loss[-2])<eta:
            break
    
    print("Final loss = ",loss[-1], ' iterations =  ',cnt)
    return w
    # End your code
    ######################################

def predictLogReg(w, x_pred):
    ######################################
    # Begin your code
    sigmoid = lambda z:1/(1 + np.exp(-z))
    
    z = np.dot(x_pred, w)
    y_pred = np.squeeze(sigmoid(z)) # np.rint()
    return y_pred
    # End your code
    ######################################
```
