<h2 style="color:rgb(0,120,170)"> Task 1: RBF kernel is indeed a kernel</h2>


The aim of this task is to give you some familiarity with kernels. We will walk you through the procedure of showing that certain functions can be used as kernels, thus defining inner products in higher dimensional spaces. Specifically, we want to do this for the most frequently used one, namely the RBF (radial basis function) kernel. This requires several preparation steps. Let us start by recalling the notion of a kernel: A map $k: X \times X \to \mathbf{R}$ is  called a kernel, if it is positive semi-definite (1) and symmetric (2), i.e.

\begin{align*}
\sum_{i,j} c_i k(\mathbf{x}^i,\mathbf{x}^j) c_j \geq 0 \quad &(1) \\
k(\mathbf{x}^i,\mathbf{x}^j)=k(\mathbf{x}^j,\mathbf{x}^i) \quad &(2)
\end{align*}

for $i=1,...,l$, $c_i \in \mathbb{R}$, $\mathbf{x}^i \in X$. An important observation is that (1) and (2) are equivalent to the statement that the gram matrix $\mathbf{K}=(k_{ij})_{i=1,\dots,l}^{j=1\dots,l}=(k(\mathbf{x}^i,\mathbf{x}^j))_{i=1,\dots,l}^{j=1\dots,l}$ is
positive semi-definite and symmetric.

Moreover, kernels have the important property that their existence is equivalent to the existence of a Hilbert space $\mathcal{H}$ and a feature mapping $\Phi: X\rightarrow\mathcal{H}$ such that $k(\mathbf{x},\mathbf{y})=\Phi(\mathbf{x})\cdot\Phi(\mathbf{y})$, where $\cdot$ denotes the inner product in $\mathcal{H}$. The important point here is that you don't have to know $\Phi$ and the feature space $\mathcal{H}$ explicitely: the most important characteristic of the Hilbert space, namely the inner product, can be specified by $k$ only. So the crucial point in order to get a suitable feature representation will be the choice of $k$.

<h3 style="color:rgb(0,120,170)">Task:</h3>

From here on out let's assume $X=\mathbb{R}^n$. The easiest kernel that all of you know for sure, is just given by the usual inner product in $\mathbf{R}^n$, i.e. $k(\mathbf{x},\mathbf{y})=\mathbf{x}^T \mathbf{y}$ for $\mathbf{x},\mathbf{y} \in X$. In the first task ensure that this is indeed a kernel by verifying (1) and (2).

<h3 style="color:rgb(208,90,80)">Calculation (10 Points):</h3>

Prrof: The kerenl is semi definite as 
\begin{align*}
\sum_{i,j}c_{i}k(\mathbf{x}^{i},\mathbf{x}^{j})c_{j}=\sum_{i,j}c_{i}(\mathbf{x}^{i})^{T}\mathbf{x}^{j}c_{j}=\sum_{i}c_{i}(\mathbf{x}^{i})^{T}\sum_{j}\mathbf{x}^{j}c_{j}=\left\Vert \sum_{i}\mathbf{x}^{i}c_{i}\right\Vert ^{2}\geq0
\end{align*}

and is symmetric as
\begin{align*}
k(\mathbf{x}^{i},\mathbf{x}^{j})=(\mathbf{x}^{i})^{T}\mathbf{x}^{j}=\sum_{l=1}^{n}\mathbf{x}_{l}^{i}\mathbf{x}_{l}^{j}=\sum_{l=1}^{n}\mathbf{x}_{l}^{j}\mathbf{x}_{l}^{i}=(\mathbf{x}^{j})^{T}\mathbf{x}^{i}=k(\mathbf{x}^{j},\mathbf{x}^{i})
\end{align*}


In applications, however, the most frequently used kernel is the Gaussian kernel, or RBF-kernel: for $\mathbf{x},\mathbf{y} \in X$, and some $\sigma \in \mathbb{R}$, $k(\mathbf{x},\mathbf{y})=\exp\!\big(\!-\!\frac{1}{2\sigma^2}\|\mathbf{x}-\mathbf{y}\|^2\big)$. Since the value of the RBF kernel decreases with distance and ranges between $0$ (in the limit) and $1$ (if $\mathbf{x}=\mathbf{y}$), it has a useful interpretation as a similarity measure. Symmetry of this kernel is clear immediately (make sure that you understand why!). Showing that this is also positive semidefinite is not as easy any more and will be, as already mentioned, the main task of this assignment. Instead of directly looking at (1), we will go a different way by ensuring the kernel property using several small steps. This procedure has the advantage that we will also prove several important general properties of kernels along the way, which turn out to be very useful in constructing new ones from already known ones.

Let us start by showing that for $\alpha \geq 0$ and a given kernel $k$, also $\alpha k$ is a kernel. We will here provide two possible proofs of this fact, so that it is hopefully easier for you to also master the remaining tasks by recognizig the available tools.
* If $\mathbf{K}$ is the Gram matrix associated to $k$, then clearly $\alpha \mathbf{K}$ is the Gram matrix associated to $\alpha k$. But then for any $\mathbf{c} \in \mathbb{R}^l$: $\mathbf{c}^T \alpha \mathbf{K} \mathbf{c}= \alpha \mathbf{c}^T  \mathbf{K} \mathbf{c} \geq 0$, since $\mathbf{K}$ is positive semidefinite and $\alpha \geq 0$. Therefore $\alpha \mathbf{K}$ is again positive semidefinite, and therefore $\alpha k$ is a kernel.
* Let $\Phi$ be the feature map associated to $k$. We will show that $\sqrt{\alpha}\,\Phi$ is the feature map associated to $\alpha k$. We calculate $\sqrt{\alpha}\,\Phi(\mathbf{x})\cdot \sqrt{\alpha}\,\Phi(\mathbf{y})=\alpha k(\mathbf{x},\mathbf{y})$ for any $\mathbf{x},\mathbf{y} \in X$

Both of these provided approaches might turn out to be useful, and it is, first of all, a good idea to make sure that you understand their meaning. With these preparation steps, you should now be able to proceed with the following task:

<h3 style="color:rgb(0,120,170)">Task:</h3>

Show that for two kernels $k$ and $k'$, $k+k'$ is again a kernel.

<h3 style="color:rgb(208,90,80)">Calculation (10 Points):</h3>
proof:

* If $\mathbf{K}$ is the Gram matrix associated to $k$ and $\mathbf{K'}$ is the Gram matrix associated to $k'$, then by considering $k'' = k+k'$, and its associated Gram matrix as $\mathbf{K''}$ it is easy to shwo that $\mathbf{K''}$ is $\mathbf{K}+\mathbf{K'}$

The Gram matrix is 
 \begin{align*}
 \mathbf{K''}=(k''_{ij})_{i=1,\dots,l}^{j=1\dots,l}=(k''(\mathbf{x}^i,\mathbf{x}^j))_{i=1,\dots,l}^{j=1\dots,l} = (k_{ij}+k'_{ij})_{i=1,\dots,l}^{j=1\dots,l}=\mathbf{K}+\mathbf{K'}
 \end{align*}
Then 
\begin{align*}
\mathbf{c}^T \mathbf{K''} \mathbf{c}= \mathbf{c}^T  (\mathbf{K}+\mathbf{K'}) \mathbf{c}= \mathbf{c}^T \mathbf{K} \mathbf{c}+\mathbf{c}^T \mathbf{K'} \mathbf{c}\geq 0
\end{align*}

Therefore, it is positive semidefinite, and therefore $k''=k+k'$ is a kernel.

Use this observation for one of the two involved Gram matrices and reformulate the resulting expression in a way such that you can use the positive semidefiniteness of the other Gram matrix.

<h3 style="color:rgb(208,90,80)">Calculation (15 Points):</h3>

We can proof by considering SVD of one Gram matrix as
\begin{align*}\mathbf{c}^{T}\mathbf{K''}\mathbf{c}=\mathbf{c}^{T}(\mathbf{K}\mathbf{K'})\mathbf{c}=\mathbf{c}^{T}(\mathbf{M}\mathbf{M}^{T}\mathbf{K'})\mathbf{c}=\sum_{i,j}c_{i}(\sum_{k}\mathbf{M}_{ik}\mathbf{M}_{jk}^{T})\mathbf{K'_{ij}}c_{j}=\sum_{k}\sum_{ij}c_{i}\mathbf{M}_{ik}\mathbf{M}_{jk}^{T}c_{j}\mathbf{K'_{ij}}=\sum_{k}\mathbf{z}_{k}^{T}\mathbf{K'}\mathbf{z}_{k}\end{align*}
where
$$\mathbf{z}_{k}=\begin{bmatrix}c_{1}\mathbf{M}_{1k}\\ \cdots \\ c_{n}\mathbf{M}_{nk}\end{bmatrix}$$
Therefore, $\mathbf{K''}$ is positive semidefinite and $k''=kk'$ is a kernel.

<h3 style="color:rgb(208,90,80)">Calculation (10 Points):</h3>

for all $x,y \in X$ we can use the continuity as 
\begin{align*}
\lim_{n\to\infty} k_n(x,y) = k(x,y)
\end{align*}
Then, 
\begin{align*}
(\forall n, \mathbf{c}^T  \mathbf{K}_{n} \mathbf{c} \geq 0) \implies  \lim_{n\to\infty} \mathbf{c}^T  \mathbf{K}_{n} \mathbf{c} =\mathbf{c}^T  \mathbf{K} \mathbf{c} \geq 0
\end{align*}

Therefore, a sequence of kernels with pointwise limit is also a kernel.

In this step we want to finalize our calculations concerning the RBF-kernel. We observe that $k'(\mathbf{x},\mathbf{y})=\exp\!\big(\!-\!\frac{1}{2\sigma^2}\|\mathbf{x}-\mathbf{y}\|^2\big)$ is just a normalized version of $k(\mathbf{x},\mathbf{y})=\exp\!\big(\frac{\mathbf{x}^T \mathbf{y}}{\sigma^2} \big)$ which we already know to be a kernel. Indeed:

\begin{align}
\frac{k(\mathbf{x},\mathbf{y})}{\sqrt{k(\mathbf{x},\mathbf{x})\,k(\mathbf{y},\mathbf{y})}}=\frac{\exp\!\big(\frac{\mathbf{x}^T \mathbf{y}}{\sigma^2} \big)}{\exp\!\big(\frac{\|\mathbf{x}\|^2}{2\sigma^2} \big)\exp\!\big(\frac{\|\mathbf{y}\|^2}{2\sigma^2} \big)}=\exp\!\big(\!-\!\frac{1}{2\sigma^2}\|\mathbf{x}-\mathbf{y}\|^2\big).
\end{align}

Therefore, we are done if we can show that kernels are also closed under normalization as above. But this is again a general fact of kernels, as you should show in the next task.
